```python
prompts = [
  "lung cancer",
  "Pancoast tumor; stage T3.",
  "bilateral IVC in otherwise normal chest CT",
  "necrotizing pneumonia on contrast-enhanced CT",
  "Moderate to severe right pleural effusion Partial right lung collapse Multiple liver masses One large liver cyst Ascites peritoneal thickening",
  "Right lung hilar mass lots of smaller metastases bilaterally mild pleural effusion pleural masses reticular opacities and interlobular septal thickening in right lung right 7th rib erosion mediastinal lymphadenopaties",
  "left rib fractures left bloody pleural effusion adjacent peripheral lung contusion small left pneumothorax",
  "right pleural based consolidation associated with mild right pleural effusion",
  "right huge basal pleural based consolidation associated with moderate right pleural effusion",
  "irregular speculated lesion in right apical region",
  "irregular speculated lesion 3 cm in right apicoposterior region",
  "enlarged right lower paratracheal lymph node. right upper lobe lung cancer spiculated 5cm.",
  "Severe right pleural effusion with subjacent passive collapse of the right lung. Collapse consolidation in the lower part of left lung lower lobe. Scattered atelectatic bands in both lungs. Tip of the central venous catheter is placed in the right atrium. No cardiomegaly, pericardial effusion and mediastinal lymphadenopathy. bilateral airways are patent.",
  "Lung spiculated cavitary lesion",
  "There is a right upper hilar mass measuring approximately 25mm",
  "Small right pleural effusion with mild bilateral atelectatic changes. Major airway tracts are patent. No pneumothorax. There is no cardiomegaly and pericardial effusion. There is also small air space opacities left lower lobe which may represent pneumonia.",
  "Thickened bilateral basilar pleural may represent mesothelioma. There are also enlarged mediastinal lymph nodes measuring 2cm in short axis. There is also moderate right effusion. Mucus plug is noted on the left main bronchus associated with atelectatic changes",
  "Patient with large R pleural effusion of unknown etiology, with suspicious findings on the L side possibly consistent with atelectasis",
  "Patient that presents with a large Type B aortic dissection with an incidentally found 2 cm LUL nodule",
  "Normal chest CT, with no presence of cardiomegaly, no presence of pericardial effusion, no presence of pleural effusion or consolidation",
  "generate an type 1 abdominal aortic aneurysm",
]
```

```python
import pandas as pd
import collections
import re

# Function to clean and tokenize a prompt.
def tokenize(text):
    # Convert text to lowercase and extract words (removing punctuation).
    return re.findall(r'\b\w+\b', text.lower())

# List to store individual prompt statistics.
stats_list = []
# Counter to accumulate overall word frequencies.
word_counts = collections.Counter()

for prompt in prompts:
    # Total number of characters (including spaces and punctuation).
    num_chars = len(prompt)
    # Tokenize the prompt.
    words = tokenize(prompt)
    # Total number of words.
    num_words = len(words)

    # Append stats for this prompt.
    stats_list.append({
        'prompt': prompt,
        'num_words': num_words,
        'num_chars': num_chars
    })

    # Update the overall word frequency.
    word_counts.update(words)

# Create a DataFrame for prompt statistics.
stats_df = pd.DataFrame(stats_list)

# Create a DataFrame for overall word counts.
word_counts_df = pd.DataFrame(list(word_counts.items()), columns=['word', 'count'])

print("Prompt Statistics DataFrame:")
print(stats_df)
print("\nOverall Word Counts DataFrame:")
print(word_counts_df)
```
